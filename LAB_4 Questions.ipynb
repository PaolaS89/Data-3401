## QUESTION 1 (10 POINTS)

### Create a line plot for the function ùë¶=2ùë•+3 for ùë• values ranging from -10 to 10. Label the x-axis as "X-axis" and the y-axis as "Y-axis". Add a title "Line Plot of y = 2x + 3".

```python
import numpy as np
import matplotlib
from matplotlib import pyplot as plt
```

```python
x = np.linspace(-10,10)
y = 2*x+3 
plt.plot(x, y) 

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Line Plot of y = 2x+3")
```

### Create a scatter plot with 100 random points. Use a different color and size for each point. Include a color bar to indicate the range of values.

```python
import numpy as np
import matplotlib 
from matplotlib import pyplot as plt
```
