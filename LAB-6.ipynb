### **Question 1**

### Create a bar chart to visualize the distribution of 'Geography' among customers.

```python
# Step 1: Bar plot for geography among customers
# Step 2: create counts of the geography via value_counts().head()
# step 3: Then we designate the figure size 
# step 4: designate the labels for the x-axis and the y-axis 
## ----> now for the x put the values and for the y do the index
# step 5: Put the titles and labels. 

N = 10 

customers_counts = df_churn['Geography'].value_counts().head(N)
plt.figure(figsize = (8, 8))
ax = sns.barplot( x = customers_counts.values,
                  y = customers_counts.index, 
                  palette = "cool")
plt.title(f"Top {N} Geography distribution among Customers")
plt.xlabel("Number of Customers")
plt.ylabel("Places")
plt.show()
```

# Observations: 

--- Most customers are from France, and the least are from Spain.  



---There is only a small difference between the number of customers from Germany and Spain.



--- The customers are only from Spain, France, and Germany

### **Question 12**

### Using the 'Churn' dataset, create a 2x2 grid of subplots. Each subplot should display one of the following:

- 1. A histogram showing the distribution of 'Age'.
- 2. A bar plot showing the count of customers in each 'Geography' category.
- 3. A scatter plot comparing 'CreditScore' and 'Balance'.
- 4. A boxplot showing the distribution of 'EstimatedSalary'.

```python

# Create a 2x2 grid of subplots
fig, axs = plt.subplots(2, 2, figsize=(15, 10))
 
# 1. Histogram showing the distribution of 'Age'
axs[0, 0].hist(df_churn['Age'], bins=20, color='skyblue', edgecolor='black')
axs[0, 0].set_title('Distribution of Age')
axs[0, 0].set_xlabel('Age')
axs[0, 0].set_ylabel('Frequency')
 
# 2. Bar plot showing the count of customers in each 'Geography' category
geography_counts = df_churn['Geography'].value_counts()
axs[0, 1].bar(geography_counts.index, geography_counts.values, color='lightgreen')
axs[0, 1].set_title('Count of Customers by Geography')
axs[0, 1].set_xlabel('Geography')
axs[0, 1].set_ylabel('Count')
 
# 3. Scatter plot comparing 'CreditScore' and 'Balance'
axs[1, 0].scatter(df_churn['CreditScore'], df_churn['Balance'], color='coral')
axs[1, 0].set_title('CreditScore vs Balance')
axs[1, 0].set_xlabel('CreditScore')
axs[1, 0].set_ylabel('Balance')
 
# 4. Boxplot showing the distribution of 'EstimatedSalary'
axs[1, 1].boxplot(df_churn['EstimatedSalary'], patch_artist=True, boxprops=dict(facecolor='lightblue'))
axs[1, 1].set_title('Distribution of EstimatedSalary')
axs[1, 1].set_ylabel('EstimatedSalary')
 
# Adjust layout
plt.tight_layout()
plt.show()

```
