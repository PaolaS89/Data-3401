```python
# Subplots with Titles 
fig, axes = plt.subplots(nrows = 1, ncols = 2)

for ax in axes: 
    ax.plot(x,y1)
    ax.set_xlabel('x')
    ax.set_ylabel('$y = x^2$')
    ax.set_title("Title")

fig.suptitle("Big Title", y= 1.1)
plt.show()
```

## Different Types of Plots

```python
x = np.linspace(0,2,50)
n = np.array([0,1,2,3,4,5])
y = np.linspace(0,1.5,50)

# plot , x, x**2 (scatter plot, line plot) and n, n**2 (axes.step() and axes.bar())
fig, axes = plt.subplots(nrows = 1, ncols = 5, figsize = (12,4))

axes[0].scatter(x,x**2)
axes[0].set_title("Scatter Plot")

axes[1].plot(x,x**2)
axes[1].set_title("Line Plot")

axes[2].step(n,n**2,lw=2,alpha=.4)
axes[2].set_title("Step Plot")

axes[3].bar(n,n**2,align='center',width=.5,alpha=.7)
axes[3].set_title("Bar Plot")

axes[4].fill_between(y,y**2,y**3)
axes[4].set_title("Fill Between")

plt.show()
```

## Plotting in Polar Coordinates

```python
fig = plt.figure()
axis = fig.add_axes([0,0,1,1], polar=True)
t = np.linspace(0,2*np.pi)
axis.plot(t,t**3,lw=2)
plt.show()
```

## 3-d Plotting

```python
## Plot f(x,y) = sin(x)+cos(y)
x = np.linspace(0,2*np.pi,100)
y = np.linspace(0,2*np.pi,100)
X,Y = np.meshgrid(x,y)

Z = np.sin(X) + np.cos(Y)
```

```python
#Creating Heatmaps

fig, axes = plt.subplots()

p = axes.pcolor(X,Y,Z)
fig.colorbar(p,ax=axes)
```

```python
# Heatmap 2

fig, axes = plt.subplots()

color = matplotlib.cm.Spectral
color = matplotlib.cm.hot

p = axes.pcolor(X,Y,Z, cmap = color)
fig.colorbar(p,ax=axes)
```

```python
fig, axes = plt.subplots()

color = matplotlib.cm.Spectral
color = matplotlib.cm.cool

p = axes.pcolor(X,Y,Z, cmap = color)
fig.colorbar(p,ax=axes)
```

```python
# Countour Plot

fig, axes = plt.subplots()

# color = matplotlib.cm.Spectral
color = matplotlib.cm.cool

contour= axes.contour(Z, cmap = color, levels = 20)
fig.colorbar(contour,ax=axes)
```
